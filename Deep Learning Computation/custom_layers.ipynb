```python
import torch
from torch import nn
```

```python
from torch import functional as F
```

```python
class customLayer(nn.Module):
    def __init__(self) -> None:
        super().__init__()

    def forward(self,X):
        return X-X.mean()
```

```python
X=torch.rand(10,30)
c_layer=customLayer()
net=nn.Sequential(nn.Linear(X.shape[1],100)
                , customLayer())
```

```python
net(X)
```

```python
class MyLinear(nn.Module):
    def __init__(self,inp,out):
        super().__init__()
        self.weight=nn.Parameter(torch.rand(inp,out))
        self.bias=nn.Parameter(torch.rand(out,))

    def forward(self,X):
        linear=torch.matmul(X,self.weight)+self.bias
        return nn.functional.relu(linear)
```

```python
net=nn.Sequential(MyLinear(X.shape[1],100)
                 , MyLinear(100,5))
```

```python
X=torch.rand(10,20)
net(X)
```
